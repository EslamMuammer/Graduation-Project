```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
employee= pd.read_csv("Employee.csv")
employee
```

```python
employee.columns
```

```python
Education_dict={
    1: "No Formal Qualifications",
    2: "High School",
    3: "Bachelors",
    4: "Masters",
    5: "Doctorate"
}

employee["Education"]=employee["Education"].replace(Education_dict)
employee["Education"]
```

```python
PerformanceRating= pd.read_csv("PerformanceRating.csv")
PerformanceRating
```

```python
Satisfaction_dict={
    1: "Very Dissatisfied",
    2: "Dissatisfied",
    3: "Neutral",
    4: "Satisfied",
    5: "Very Satisfied"
}

PerformanceRating["EnvironmentSatisfaction"]= PerformanceRating["EnvironmentSatisfaction"].replace(Satisfaction_dict)
PerformanceRating["JobSatisfaction"]= PerformanceRating["JobSatisfaction"].replace(Satisfaction_dict)
PerformanceRating["RelationshipSatisfaction"]= PerformanceRating["RelationshipSatisfaction"].replace(Satisfaction_dict)
PerformanceRating["WorkLifeBalance"]= PerformanceRating["WorkLifeBalance"].replace(Satisfaction_dict)
```

```python
Rating_dict= {
    1: "Unacceptable",
    2: "Needs Improvement",
    3: "Meets Expectation",
    4: "Exceeds Expectation",
    5: "Above and Beyond" 
}

PerformanceRating["SelfRating"]= PerformanceRating["SelfRating"].replace(Rating_dict)
PerformanceRating["ManagerRating"]= PerformanceRating["ManagerRating"].replace(Rating_dict)
```

```python
PerformanceRating
```

```python
df= pd.merge(employee, PerformanceRating, on= "EmployeeID")
df
```

```python
average_distance_by_department= df.groupby("Department")["DistanceFromHome (KM)"].mean()
average_distance_by_department
```

```python
average_distance_by_satisfaction= df.groupby("JobSatisfaction")["DistanceFromHome (KM)"].mean()
average_distance_by_satisfaction
```

```python
average_distance_by_salary= df.groupby("Salary")["DistanceFromHome (KM)"].mean().reset_index()
average_distance_by_salary
```

```python
# How does average distance from home to work by department affect employee satisfaction?

colors_department = ["rosybrown", "darkgray", "salmon"]  
colors_satisfaction = ["lightgreen", "lightblue", "teal", "khaki", "darkseagreen"]

fig, axes= plt.subplots(nrows= 1, ncols= 2)
axes[0].bar(average_distance_by_department.index, average_distance_by_department.values, color=colors_department)
axes[0].set_xticks(range(len(average_distance_by_department.index)))
axes[0].set_xlabel("Average Distance By Department")
axes[0].set_xticklabels(average_distance_by_department.index, rotation=90)

axes[1].bar(average_distance_by_satisfaction.index, average_distance_by_satisfaction.values, color=colors_satisfaction)
axes[1].set_xticks(range(len(average_distance_by_satisfaction.index)))
axes[1].set_xlabel("Average Distance By Satisfaction")
axes[1].set_xticklabels(average_distance_by_satisfaction.index, rotation=90)
plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 8))  

# How does overtime affect employees' job satisfaction levels?
sns.boxplot(x="OverTime", y="JobSatisfaction", data=df, ax=axs[0, 0], palette='Set2', hue="OverTime")
axs[0, 0].set_title("Job Satisfaction by Overtime")
axs[0, 0].set_xlabel("Overtime")
axs[0, 0].set_ylabel("Job Satisfaction")

# How does overtime affect employees' work-life balance?
sns.boxplot(x="OverTime", y="WorkLifeBalance", data=df, ax=axs[0, 1], palette='Set2', hue="OverTime")
axs[0, 1].set_title("Work-Life Balance by Overtime")
axs[0, 1].set_xlabel("Overtime")
axs[0, 1].set_ylabel("Work-Life Balance")

# Does overtime work affect the number of years of service in the company?
sns.boxplot(x="OverTime", y="YearsAtCompany", data=df, ax=axs[1, 0], palette='Set2', hue="OverTime")
axs[1, 0].set_title("Years at Company by Overtime")
axs[1, 0].set_xlabel("Overtime")
axs[1, 0].set_ylabel("Years At Company")

# How is salary distributed between employees who work overtime and those who do not?
sns.boxplot(x="OverTime", y="Salary", data=df, ax=axs[1, 1], palette='Set2', hue="OverTime")
axs[1, 1].set_title("Salary By Overtime")
axs[1, 1].set_xlabel("Over Time")
axs[1, 1].set_ylabel("Salary")

plt.tight_layout()   
plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 10))   

# How is the salary level distributed among all employees?
axs[0, 0].hist(df["Salary"], bins=20, color='skyblue', edgecolor='black')
axs[0, 0].set_title("Salary Distribution among Employees")
axs[0, 0].set_xlabel("Salary")
axs[0, 0].set_ylabel("Number of Employees")

# Are there noticeable differences in salaries between different departments?
df.boxplot(column="Salary", by="Department", ax=axs[0, 1], grid=False, patch_artist=True)
axs[0, 1].set_title("Salary Distribution by Department")
axs[0, 1].set_xlabel("Department")
axs[0, 1].set_ylabel("Salary")
axs[0, 1].tick_params(axis='x', rotation=45)

# How does the level of education affect the salaries granted to employees?
df.boxplot(column="Salary", by="Education", ax=axs[1, 0], grid=False, patch_artist=True)
axs[1, 0].set_title("Salary Distribution by Education Level")
axs[1, 0].set_xlabel("Education Level")
axs[1, 0].set_ylabel("Salary")
axs[1, 0].tick_params(axis='x', rotation=45)

# How do average salaries differ between different departments?
average_salary_by_department = df.groupby("Department")["Salary"].mean()
average_salary_by_department.plot(kind="bar", color="orange", ax=axs[1, 1])
axs[1, 1].set_title("Average Salary by Department")
axs[1, 1].set_xlabel("Department")
axs[1, 1].set_ylabel("Average Salary")
axs[1, 1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
df['TrainingTaken'] = df['TrainingOpportunitiesTaken'].apply(lambda x: 1 if x > 0 else 0)

fig, axs = plt.subplots(2, 2, figsize=(12, 10))  

# How does the level of education affect the rate at which employees benefit from training opportunities?
training_rate_by_education = df.groupby("Education")["TrainingTaken"].mean() * 100
training_rate_by_education.plot(kind="bar", ax=axs[0, 0], color='skyblue')
axs[0, 0].set_title("Training Opportunities Taken Rate by Education Level")
axs[0, 0].set_xlabel("Education Level")
axs[0, 0].set_ylabel("Training Opportunities Taken Rate (%)")
axs[0, 0].set_ylim(0, 100)
axs[0, 0].grid(axis="y")

# Are there differences in the rate of utilization of training between different departments?
training_rate_by_department = df.groupby("Department")["TrainingTaken"].mean() * 100
training_rate_by_department.plot(kind="bar", ax=axs[0, 1], color='lightgreen')
axs[0, 1].set_title("Training Opportunities Taken Rate by Department")
axs[0, 1].set_xlabel("Department")
axs[0, 1].set_ylabel("Training Opportunities Taken Rate (%)")
axs[0, 1].set_ylim(0, 100)
axs[0, 1].grid(axis="y")

# How does overtime work affect the benefit from training opportunities?
training_rate_by_overtime = df.groupby("OverTime")["TrainingTaken"].mean() * 100
training_rate_by_overtime.plot(kind="bar", ax=axs[1, 0], color='salmon')
axs[1, 0].set_title("Training Opportunities Taken Rate by Overtime")
axs[1, 0].set_xlabel("Overtime")
axs[1, 0].set_ylabel("Training Opportunities Taken Rate (%)")
axs[1, 0].set_ylim(0, 100)
axs[1, 0].grid(axis="y")

axs[1, 1].axis('off')
plt.tight_layout()
plt.show()
```

```python
df["Promoted"] = df["YearsInMostRecentRole"] > df["YearsWithCurrManager"]
promotion_counts = df["Promoted"].value_counts().rename({True: 'Promoted', False: 'Not Promoted'})

fig, axs = plt.subplots(1, 2, figsize=(14, 6))

# How many employees received a promotion compared to those who did not?
sns.barplot(x=promotion_counts.index, y=promotion_counts.values, palette='viridis', hue=promotion_counts.index,ax=axs[0])
axs[0].set_title("Internal Promotion Rate")
axs[0].set_xlabel("Promotion Status")
axs[0].set_ylabel("Number of Employees")
axs[0].set_xticks([0, 1])
axs[0].set_xticklabels(["Promoted", "Not Promoted"])

# How are promotions distributed among employees?
axs[1].pie(promotion_counts, labels=promotion_counts.index, autopct='%1.1f%%', startangle=90, colors=['#66c2a5', '#fc8d62'])
axs[1].axis('equal')    
axs[1].set_title("Distribution of Promotions")

plt.tight_layout()

plt.show()
```

```python
training= pd.crosstab(df["TrainingOpportunitiesTaken"], df["TrainingOpportunitiesWithinYear"])
training     
```

```python
fig, axs = plt.subplots(1, 2, figsize=(14, 6))

# How is the manager's evaluation distributed among employees?
sns.histplot(df["ManagerRating"], bins=5, kde=True, color='skyblue', ax=axs[0])
axs[0].set_title("Distribution of Manager Rating")
axs[0].set_xlabel("Manager Rating")
axs[0].set_ylabel("Number of Employees")
axs[0].tick_params(axis="x", rotation=45)

# How does manager evaluation affect employee satisfaction?
sns.boxplot(x="ManagerRating", y="JobSatisfaction", data=df, ax=axs[1], palette='coolwarm', hue="ManagerRating")
axs[1].set_title("Job Satisfaction by Manager Rating")
axs[1].set_xlabel("Manager Rating")
axs[1].set_ylabel("Job Satisfaction")

plt.tight_layout()
plt.show()
```

```python
TrainingSatisfactionRate = df["TrainingOpportunitiesTaken"] / df["TrainingOpportunitiesWithinYear"]

# How is the employee satisfaction rate distributed regarding the available training opportunities?
fig, ax = plt.subplots(figsize=(10, 6))
sns.histplot(TrainingSatisfactionRate, bins=10, kde=True, color='purple')

ax.set_title("Training Satisfaction Rate Distribution")
ax.set_xlabel("Training Satisfaction Rate")
ax.set_ylabel("Number of Employees")

plt.tight_layout()
plt.show()
```

```python
satisfaction_mapping = {
    "Very Dissatisfied": 1,
    "Dissatisfied": 2,
    "Neutral": 3,
    "Satisfied": 4,
    "Very Satisfied": 5
}

temp_job_satisfaction = df["JobSatisfaction"].copy().map(satisfaction_mapping)
temp_environment_satisfaction = df["EnvironmentSatisfaction"].copy().map(satisfaction_mapping)
temp_work_life_balance = df["WorkLifeBalance"].copy().map(satisfaction_mapping)
temp_relationship_satisfaction = df["RelationshipSatisfaction"].copy().map(satisfaction_mapping)

df["OverallSatisfactionRate"] = (temp_job_satisfaction + 
                                  temp_environment_satisfaction + 
                                  temp_work_life_balance + 
                                  temp_relationship_satisfaction) / 4


# How is the overall employee satisfaction rate with the work environment in general distributed?
fig, ax = plt.subplots(figsize=(10, 6))
sns.histplot(df["OverallSatisfactionRate"], bins=10, kde=True, color='orange', ax=ax)

ax.set_title("Overall Satisfaction Rate Distribution")
ax.set_xlabel("Overall Satisfaction Rate")
ax.set_ylabel("Number of Employees")

plt.tight_layout()
plt.show()
```
